```python
email= input("enter the email:")
if email==""  :
    print('email should not be blank')
elif  not('.'in email and  '@' in email):
    print('email must contain \'.\' and \'@\'')
elif len(email) >254:
    print('email should not be more than 254')
elif email[0].isalnum() and email[-1].isalnum:
    print("email should be start and ends with the string and integer")
elif email.count('@') !=1:
    print("it contains more @")
else :
    print('email is valid') 
```
